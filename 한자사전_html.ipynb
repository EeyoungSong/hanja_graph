```python
from bs4 import BeautifulSoup
```

```python
# BeautifulSoup을 사용하여 HTML을 파싱합니다.
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 한자와 뜻을 담고 있는 태그를 추출합니다.
div_tag_hanja = soup.findAll('div', class_='hanja_word')

h_arr = []
m_arr = []
for tag_content in div_tag_hanja:
    h = tag_content.find('a')
    m = tag_content.find('div', class_='mean')
    try:
        print(h.text.strip(), m.text.strip())
        h_arr.append(h.text.strip())
        m_arr.append(m.text.strip())
    except:
        print("error")

# 모양자를 담고 있는 태그를 추출합니다.
div_tag_shape = soup.findAll('div', class_='info_item shape')

s_arr = []
for tag_content in div_tag_shape:
     s = tag_content.find('div', class_="desc", lang='zh')
     print(s.text)
     s_arr.append(s.text)

print(h_arr)
print(m_arr)
```

```python
# 모양자를 쪼개서 리스트 만들고, 모양자 중복 제거 리스트 만드는 코드
s_split_list = []
s_split_list_all = []
for s in s_arr:
    s_split_list.append(s.split(" + "))
    s_split_list_all.extend(s.split(" + "))

s_dedu = list(set(s_split_list_all))
print(s_split_list)
print(s_split_list_all)
print(s_dedu)

# 모양자 한자랑 의미 쪼개는 코드
moyangja_list = []
moyangja_mean_list = []
for s in s_split_list:
    for t in s:
        a = t.split("(")[0]
        b = t.split("(")[1].rstrip(")")
        moyangja_list.append(a)
        moyangja_mean_list.append(b)
print(moyangja_list)
print(moyangja_mean_list)

#split list에서 괄호 제거
sub_split_list = []
for s in s_split_list:
    temp = [t.split("(")[0] for t in s]
    sub_split_list.append(temp)


# 한자에 해당하는 모양자 매핑 시키는 코드
f_t_arr = []
for s, h in zip(sub_split_list, h_arr):
    for t in s:
        f_t_arr.append([t, h])
print(f_t_arr)

```

```python
# 중복 제거한 모양자 리스트 쪼개기
dedu_hanja = []
dedu_mean = []
for s in s_dedu:
    dedu_hanja.append(s.split("(")[0])
    dedu_mean.append(s.split("(")[1].rstrip(")"))
print(dedu_hanja)
print(dedu_mean)
```

```python
import pandas
# 한자 리스트를 엑셀로 변환하여 저장합니다.
list_df = pandas.DataFrame({"hanja": h_arr, "mean": m_arr})
pandas_to_excel = list_df.to_excel('crolling_hanja_mean.xlsx')
```

```python
# 모양자 리스트를 엑셀로 변환하여 저장합니다.
list_df = pandas.DataFrame({"hanja": dedu_hanja, "mean": dedu_mean})
pandas_to_excel = list_df.to_excel('crolling_moyangja_mean.xlsx')
```

```python
# 중복되는 모양자와 한자들을 제거합니다.

arr = m_arr + dedu_mean

test_dic = {}
for elem in arr:
    if elem in test_dic:
        test_dic[elem] += 1
    else:
        test_dic[elem] = 1

for key, value in test_dic.items():
    if value > 1:
        print(key, value)
```

```python
# 한자와 한자의 모양자를 매핑한 리스트를 엑셀로 변환합니다.

list_df = pandas.DataFrame(f_t_arr, columns=['shape', 'hanja'])
print(list_df)
pandas_to_excel = list_df.to_excel('crolling_hanja.xlsx')
```
